```python
import pandas as pd

data = {
    "State": ["Tamil Nadu", "Kerala", "Karnataka", "Andhra Pradesh", "Telangana"],
    "Year": [2020, 2020, 2020, 2020, 2020],
    "Unemployment_Rate (%)": [7.5, 9.2, 6.8, 8.1, 7.9],
    "Employed_Population": [35000000, 15000000, 32000000, 28000000, 22000000]
}

df = pd.DataFrame(data)
df.to_csv('unemployment_sample_data.csv', index=False)

df.head()
```

```python
import pandas as pd

df = pd.read_csv('unemployment_sample_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(df["State"], df["Unemployment_Rate (%)"])
plt.title("Unemployment Rate by State")
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[["Employed_Population"]]
y = df["Unemployment_Rate (%)"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

print("Predictions:", predictions)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(df["State"], df["Unemployment_Rate (%)"])

plt.title("Unemployment Rate by State")
plt.xlabel("State")
plt.ylabel("Unemployment Rate (%)")

plt.xticks(rotation=45)
plt.show()
```
